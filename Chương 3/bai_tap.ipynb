```python
import numpy as np 
import time 
# Tạo một mảng NumPy với 100.000.000 phần tử 
arr = np.arange(100000000, dtype=np.int64) 
# Tính tổng các phần tử của mảng bằng vòng lặp 
start_time = time.time() 
total = np.int64(0) 
for num in arr: 
    total += num 
end_time = time.time() 
elapsed_time_loop = end_time - start_time 
print("Tổng các phần tử mảng arr khi sử dụng vòng lặp:", total) 
print("Thời gian (giây) khi sử dụng vòng lặp:", elapsed_time_loop) 
# Tính tổng các phần tử của mảng bằng np.sum() 
start_time = time.time() 
total = np.sum(arr) 
end_time = time.time() 
elapsed_time_npsum = end_time - start_time 
print("Tổng các phần tử mảng arr khi sử dụng hàm np.sum():", total) 
print("Thời gian (giây) khi sử dụng np.sum():", elapsed_time_npsum) 
num_times_faster = elapsed_time_loop / elapsed_time_npsum 
print("Sử dụng vòng lặp chậm hơn: ~",int(num_times_faster),"lần") 
```

```python
import numpy as np 
import timeit # Import thư viện timeit để đo thời gian thực thi của các đoạn code. 
 
# Tính tổng của mảng sử dụng Numpy 
np_setup = "import numpy as np; np_array = np.arange(1, 1000001)" 
np_code  = "np_sum = np.sum(np_array)" 
np_time  = timeit.timeit(stmt=np_code, setup=np_setup, number=10) 
 
# Tính tổng của danh sách sử dụng List 
list_setup = "list_array = list(range(1, 1000001))" 
list_code = "list_sum = sum(list_array)" 
list_time = timeit.timeit(stmt=list_code, setup=list_setup, 
number=10) 
 
print("Thời gian tính toán của Numpy:", np_time) 
print("Thời gian tính toán của List:", list_time) 
 
print("Numpy tính nhanh hơn List ở lượt chạy là ~ ", int(list_time/np_time), "lần!!")
```

```python
import numpy as np 
 
arr1 = np.array([1, 2, 3]) 
arr2 = np.array([4, 5, 6]) 
 
#Phép Cộng:Sử dụng hàm np.add() hoặc toán tử '+'. 
Tong = np.add(arr1,arr2) 
#hoặc tong = arr1+arr2 
 
#Phép Trừ: Sử dụng hàm np.subtract() hoặc toán tử '-'. 
Hieu = np.subtract(arr1,arr2) 
#hoặc Tru = arr1-arr2 
 
#Phép Nhân: Sử dụng hàm np.multiply() hoặc toán tử '*' 
Tich = np.multiply(arr1,arr2) 
#Hoặc arr1*arr2 
 
#Phép Chia: Sử dụng hàm np.divide() hoặc toán tử '/' 
Thuong = np.divide(arr1,arr2) 
# Về sin cos tan, mp.sin(), np.cos(), np.tan, np.log(), np.log()^2,np.sprt()
 
print("Tổng của arr1 + arr2 là:",Tong) 
print("Hiệu của arr1 - arr2 là:",Hieu) 
print("Tích của arr1 * arr2 là:",Tich) 
print("Thương của arr1/arr2 là:",Thuong)
```

```python
# 15. less – nhỏ hơn (<)
np.less(a1, a2)   # ➜ array([ True, True, True])

# 16. less_equal – nhỏ hơn hoặc bằng (<=)
np.less_equal(a2, a1)   # ➜ array([False, False, False])

# 17. equal – bằng nhau (==)
np.equal(a1, a2)   # ➜ array([False, False, False])

# 18. not_equal – khác nhau (!=)
np.not_equal(a1, a2)   # ➜ array([ True, True, True])

# Các phép logic (với mảng Boolean)
a = np.array([True, True, False])
b = np.array([True, False, False])

# 19. logical_and – Phép “và” logic
np.logical_and(a, b)   # ➜ array([ True, False, False])

# 20. logical_or – Phép “hoặc” logic
np.logical_or(a, b)    # ➜ array([ True, True, False])

# 21. logical_xor – Phép “hoặc loại trừ” logic
np.logical_xor(a, b)   # ➜ array([False, True, False])
```

```python
import numpy as np 
arr = np.array([1, 2, 3, 4, 5]) 
 
# Căn bậc hai của các phần tử trong mảng 
kq_can_hai = np.sqrt(arr)  
 
# Lũy thừa 2 của các phần tử trong mảng 
kq_luy_thua_2 = np.power(arr, 2) 
 
# Logarit tự nhiên (cơ số e) của các phần tử trong mảng 
kq_Logarit = np.log(arr) 
  
# Giá trị tuyệt đối của các phần tử trong mảng 
kq_tri_tuyet_doi = np.abs(arr) 
 
# Hàm logic NOT (đảo ngược giá trị) của các phần tử trong mảng 
kq_not_logic = np.logical_not(arr) 
#Hàm sin của các phần tử trong mảng 
kq_sin=np.sin(arr) 
print("Căn bậc hai của từng phần tử trong mảng là: ", kq_can_hai)  
print("Lũy thừa bậc hai của từng phần tử trong mảng là: ", kq_luy_thua_2) 
print("Logarit của từng phần tử trong mảng là: ", kq_Logarit)  
print("Trị tuyệt đối của từng phần tử trong mảng là: ", kq_tri_tuyet_doi)  
print("Kết quả đảo ngược giá trị của các phần tử trong mảng là:", kq_not_logic) 
print("Kết quả hàm sin của các phần tử trong mảng là:", kq_sin)
```

```python
import numpy as np 
# Tạo mảng dữ liệu với kích thước lớn (1000 phần tử) 
data = np.random.randint(100, 200, size=1000) 
# Tạo một giá trị trung bình (số thực) 
mean_value = np.array([data.mean()]) 
# Tính tỷ lệ phần trăm tăng trưởng bằng cách áp dụng broadcasting 
percentage_growth= (data - mean_value) / mean_value * 100 
# In ra 10 phần tử đầu tiên của mảng dữ liệu và tỷ lệ phần trăm tăng trưởng 
print("Dữ liệu gốc:", data[:10]) 
print("Tỷ lệ phần trăm tăng trưởng:", 
percentage_growth[:10])
```
