```python
import math
class PS : 
    def __init__(self, tu_so = 0, mau_so = 1):
        self.tu_so = tu_so
        self.mau_so = mau_so
    def nhap_so(self) :
        self.tu_so = int(input("Nhập tử số của bạn là: "))
        self.mau_so  = int(input("Nhập mẫu số của bạn là: "))
        while self.mau_so == 0 :
            print("Mấu số ko hợp lệ")
            self.mau_so = int(input("Nhập lại mẫu số của bạn: "))
    def rut_gon(self):
        ucln = math.gcd(self.tu_so, self.mau_so)
        tu = self.tu_so // ucln
        mau = self.mau_so // ucln
        return tu, mau
    def tinh_toan(self):
        return self.tu_so / self.mau_so
    def in_so(self):
        print(f"Phân số của bạn là : {self.tu_so} / {self.mau_so}")
        tu, mau = self.rut_gon()
        print(f"Phân số sau khi rút gọn là:  {tu} / {mau}")
ps = PS()
ps.nhap_so()
ps.in_so()
```

```python
import math
class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def display(self):
        print(f"({self.x}, {self.y})")
class Elip(Diem):
    def __init__(self, x=0, y=0, a=1, b=1):
        Diem.__init__(self, x, y)   
        self.a = a 
        self.b = b  
    def dien_tich(self):
        return math.pi * self.a * self.b
class DuongTron(Elip):
    def __init__(self, x=0, y=0, r=1):
        Elip.__init__(self, x, y, r, r)   
        self.r = r
    def dien_tich(self):
        return math.pi * (self.r ** 2)
print("Nhập thông tin elip:")
x = float(input("Tọa độ tâm x: "))
y = float(input("Tọa độ tâm y: "))
a = float(input("Bán trục lớn a: "))
b = float(input("Bán trục nhỏ b: "))
elip = Elip(x, y, a, b)
print(f"Diện tích elip = {elip.dien_tich():.2f}")
print("Nhập thông tin đường tròn:")
x = float(input("Tọa độ tâm x: "))
y = float(input("Tọa độ tâm y: "))
r = float(input("Bán kính r: "))
dt = DuongTron(x, y, r)
print(f"Diện tích đường tròn = {dt.dien_tich():.2f}")
```

```python
import math
class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def chu_vi(self):
        return self.a + self.b + self.c
    def dien_tich(self):
        p = self.chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
class TamGiacVuong(TamGiac):
    def __init__(self, a, b, c):
        TamGiac.__init__(self, a, b, c)
    def la_tam_giac_vuong(self):
        canh = sorted([self.a, self.b, self.c])
        return math.isclose(canh[0]**2 + canh[1]**2, canh[2]**2)
class TamGiacCan(TamGiac):
    def __init__(self, a, b, c):
        TamGiac.__init__(self, a, b, c)
    def la_tam_giac_can(self):
        return self.a == self.b or self.b == self.c or self.a == self.c
class TamGiacDeu(TamGiacCan):
    def __init__(self, a):
        TamGiacCan.__init__(self, a, a, a)
    def la_tam_giac_deu(self):
        return self.a == self.b == self.c
    
tg = TamGiac(3, 4, 5)
print("Chu vi TG:", tg.chu_vi())
print("Diện tích TG:", round(tg.dien_tich(), 2))
tg_vuong = TamGiacVuong(3, 4, 5)
print("Tgiac vuông 3-4-5 có phải vuông không - ", tg_vuong.la_tam_giac_vuong())
tg_can = TamGiacCan(5, 5, 8)
print("Tgiac cân 5-5-8 có phải cân không - ", tg_can.la_tam_giac_can())
tg_deu = TamGiacDeu(6)
print("Tgiac đều cạnh 6 có phải đều không - ", tg_deu.la_tam_giac_deu())
print("Chu vi:", tg_deu.chu_vi())
print("Diện tích:", round(tg_deu.dien_tich(), 2))

```
