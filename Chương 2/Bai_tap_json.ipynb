```python
import json
# Chuỗi JSON
chuoi_json = '{"ten": "An", "tuoi": 20, "que": "Ha Noi"}'
# Chuyển chuỗi JSON thành đối tượng Python (từ điển)
doi_tuong = json.loads(chuoi_json)
print("Kết quả sau khi chuyển đổi:")
print(doi_tuong)
print("Tên:", doi_tuong["ten"])
```

```python
import json
# Tạo từ điển Python
tu_dien = {"c": 3, "a": 1, "b": 2}
# Chuyển đổi sang JSON (sắp xếp theo khóa, thụt lề 4)
chuoi_json = json.dumps(tu_dien, sort_keys=True, indent=4)
print("Kết quả JSON:")
print(chuoi_json)
```

```python
import json
# Đọc dữ liệu JSON
with open("bai2_10.json", "r", encoding="utf-8") as tep:
    du_lieu = json.load(tep)
tong_nv = sum(dv["so_nv"] for dv in du_lieu["don_vi"])
print("Tên công ty:", du_lieu["ten_cong_ty"])
print("Địa chỉ:", du_lieu["dia_chi"])
print("\n--- Thống kê nhân viên theo đơn vị ---")
for i, dv in enumerate(du_lieu["don_vi"], 1):
    ti_le = dv["so_nv"] / tong_nv * 100
    print(f"{i}. Tên đơn vị: {dv['ten_don_vi']}")
    print(f"   Số nhân viên: {dv['so_nv']}")
    print(f"   Tỷ lệ: {ti_le:.2f}%")
```

```python
import json
import datetime
danh_sach_giao_dich = []
while True:
    print("\n--- NHẬP GIAO DỊCH ---")
    loai = input("Nhập loại giao dịch (tiền/vàng): ")
    so_tien = float(input("Nhập số tiền: "))
    noi_dung = input("Nội dung: ")
    danh_sach_giao_dich.append({
        "loai": loai,
        "so_tien": so_tien,
        "noi_dung": noi_dung
    })
    tiep = input("Nhập thêm giao dịch? (y/n): ")
    if tiep.lower() != "y":
        break
ghi = input("Bạn có muốn ghi vào tập tin không? (1: Có / 0: Không): ")
if ghi == "1":
    # Lấy ngày giờ hiện tại
    thoi_gian = datetime.datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    ten_tap_tin = f"{thoi_gian}.json"
    with open(ten_tap_tin, "w", encoding="utf-8") as tep:
        json.dump(danh_sach_giao_dich, tep, ensure_ascii=False, indent=4)
    print("Đã ghi dữ liệu vào tập tin:", ten_tap_tin)
else:
    print("Không ghi dữ liệu vào tập tin.")
```

```python
import requests  # Thư viện chuyên để gọi dữ liệu từ Internet
# Địa chỉ API chứa danh sách các bài post
api_url = "https://jsonplaceholder.typicode.com/posts"
# Gửi yêu cầu (request) đến API và lấy dữ liệu
phan_hoi = requests.get(api_url)
# Chuyển dữ liệu JSON thành danh sách đối tượng Python
danh_sach_post = phan_hoi.json()
# In ra tổng số bài post
print("Tổng số bài post:", len(danh_sach_post))
print("-" * 60)
# In ra từng bài post
for post in danh_sach_post:
    print(f"userId: {post['userId']}")
    print(f"id: {post['id']}")
    print(f"title: {post['title']}")
    print(f"body: {post['body']}")
    print("-" * 60)
```

```python
import requests
# Địa chỉ API của bài post nổi bật
api_url = "https://jsonplaceholder.typicode.com/comments?postId=1"
# Gửi yêu cầu đến API và lấy dữ liệu
phan_hoi = requests.get(api_url)
danh_sach_noi_bat = phan_hoi.json()
print(" Danh sách các bài post nổi bật (giới hạn 3 bài):")
print("-" * 60)
# Duyệt 3 phần tử đầu tiên
for bai in danh_sach_noi_bat[:3]:
    print(f"ID: {bai['id']}")
    print(f"Tên người bình luận: {bai['name']}")
    print(f"Email: {bai['email']}")
    print(f"Nội dung: {bai['body']}")
    print("-" * 60)
```
